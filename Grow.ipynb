```python
from ipywidgets import widgets
from IPython.display import display
  
from IPython.display import clear_output
```

```python

service_button = widgets.RadioButtons(
    options=['CO-OP', 'CULIANCE', 'BOTH'],
    value='BOTH',
    description='Service to keep:',
    disabled=False,
    style = {'description_width': 'initial'}
)
```

```python
both_services=widgets.IntSlider(min=0, max=100,value = 0,style = {'description_width': 'initial'})
```

```python
display(both_services)

```

```python
both_services.value
```

```python
def whatif(service):
    n_coop = 125376
    n_culiance = 79429
    n_visa = 29633
    if(service == 'CULIANCE'):
        n_culiance = n_culiance + (both_services.value/100) * n_coop
        n_visa = n_visa + ((100-both_services.value)/100)*n_coop
        return n_culiance,n_visa
    elif(service == 'CO-OP'):
        
        n_visa = n_visa + n_culiance
        return n_culiance,n_visa
    else:
        return 
```

```python
def disp():
#    clear_output()
    service_name = service_button.value
    
    if service_name == 'CO-OP' or service_name == 'CULIANCE':
        n_ser,n_v = whatif(service_button.value)
        print("Number of "+str(service_button.value)+"transactions: "+str(n_ser))
        print("Number of VISA transactions: "+str(n_v))
        t_loss = (n_coop*0.07)+(n_culiance*0.03)
        t_profit = n_visa*4
    else:
        print("Number of CO-OP transactions: "+str(n_coop))
        print("Number of CULIANCE transactions: "+str(n_culiance))
        print("Number of VISA transactions: "+str(n_visa))
#        print("Total Loss: "+str((n_coop*0.07)+(n_culiance*0.03)))
#        print("Total Profit: "+ str(n_visa*4))
        t_loss = (n_coop*0.07)+(n_culiance*0.03)
        t_profit = n_visa*4
    if service_name == 'CULIANCE':
        print("Total Loss: "+str(n_ser*0.03))
        print("Total Profit: "+ str(n_v*4))
        t_loss = (n_coop*0.07)+(n_culiance*0.03)
        t_profit = n_visa*4
    elif service_name == 'CO-OP':
        print("Total Loss: "+str(n_ser*0.07))
        print("Total Profit: "+ str(n_v*4))
        t_loss = (n_coop*0.07)+(n_culiance*0.03)
        t_profit = n_visa*4
        
    

    
        
```
